# Importar dados

```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

# Definição do classificador

```python
model = RandomForestClassifier(n_estimators=100, n_jobs=-1, random_state=0)
```

# Seleção e padronização de parâmetros

```python
def sex_transform(value):
    if value == 'female':
        return 1
    else:
        return 0
```

```python
train['Binary_sex'] = train['Sex'].map(sex_transform)
```

```python
variables = ['Binary_sex', 'Age']
```

```python
x = train[variables].fillna(-1)
y = train['Survived']
```

# Divisão dos conjuntos de dados

```python
np.random.seed(0)
x_train, x_valid, y_train, y_valid = train_test_split(x, y, test_size=0.5)
```

# Treinamento do modelo

```python
model.fit(x_train, y_train)
```

# Predição do conjunto de validação

```python
p = model.predict(x_valid)
```

```python
np.mean(y_valid == p)
```

Caso no qual todos os passageiros do sexo feminino sobreviveram

```python
p = (x_valid['Binary_sex'] == 1).astype(np.int64)
np.mean(y_valid == p)
```

# Validação cruzada

```python
from sklearn.model_selection import RepeatedKFold
```

```python
results = []
kf = RepeatedKFold(n_splits=2, n_repeats=10, random_state=10)

for train_rows, valid_rows in kf.split(x):
    #print('Train:', train_rows.shape[0])
    #print('Valid:', valid_rows.shape[0])

    x_train, x_valid = x.iloc[train_rows], x.iloc[valid_rows]
    y_train, y_valid = y.iloc[train_rows], y.iloc[valid_rows]

    model = RandomForestClassifier(n_estimators=100, n_jobs=-1, random_state=0)
    model.fit(x_train, y_train)

    p = model.predict(x_valid)

    acc = np.mean(y_valid == p)
    results.append(acc)
    #print('Acc:', acc)

    #print()
```

```python
np.mean(results)
```

# Adição de variáveis

```python
variables = ['Binary_sex', 'Age', 'Pclass', 'SibSp', 'Parch', 'Fare']
```

```python
x = train[variables].fillna(-1)
y = train['Survived']
```

```python
results = []
kf = RepeatedKFold(n_splits=2, n_repeats=10, random_state=10)

for train_rows, valid_rows in kf.split(x):
    #print('Train:', train_rows.shape[0])
    #print('Valid:', valid_rows.shape[0])

    x_train, x_valid = x.iloc[train_rows], x.iloc[valid_rows]
    y_train, y_valid = y.iloc[train_rows], y.iloc[valid_rows]

    model = RandomForestClassifier(n_estimators=100, n_jobs=-1, random_state=0)
    model.fit(x_train, y_train)

    p = model.predict(x_valid)

    acc = np.mean(y_valid == p)
    results.append(acc)
    #print('Acc:', acc)

    #print()
```

```python
np.mean(results)
```
