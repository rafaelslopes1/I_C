# Importar dados

```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

# Definição do classificador

```python
model = RandomForestClassifier(n_estimators=100, n_jobs=-1, random_state=0)
```

# Seleção e padronização de parâmetros

```python
def sex_transform(value):
    if value == 'female':
        return 1
    else:
        return 0
```

```python
variables = ['Binary_sex', 'Age']
```

```python
train['Binary_sex'] = train['Sex'].map(sex_transform)
test['Binary_sex'] = test['Sex'].map(sex_transform)
```

```python
x = train[variables].fillna(-1)
y = train['Survived']
```

```python
x_pred = test[variables].fillna(-1)
```

# Treino

```python
model.fit(x, y)
p = model.predict(x_pred)
```

```python
sub = pd.Series(p, index=test['PassengerId'], name='Survived')
```

```python
sub.to_csv('submission.csv', header=True)
```
